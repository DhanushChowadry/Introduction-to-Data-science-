# Step-5:Finding Null-Values

```python
data.isnull().sum()
```

# Step-6:Filling the missing values(Categorical&Numerical)

```python
data['director'].fillna(data['director'].mode()[0],inplace=True)
data['cast'].fillna(data['cast'].mode()[0],inplace=True)
data['country'].fillna(data['country'].mode()[0],inplace=True)
data['date_added'].fillna(data['date_added'].mode()[0],inplace=True)
data['rating'].fillna(data['rating'].mode()[0],inplace=True)
data['duration'].fillna(data['duration'].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

# Step-7:Visualization Techniques

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Basic Plot

```python
data.head()
```

```python
#Lineplot
sns.lineplot(x='release_year',y='show_id',data=data)
plt.title('Lineplot')
plt.show()
```

```python
#Bar plot
sns.barplot(x='release_year',y='show_id',data=data)
plt.title('Barplot')
plt.show()
```

```python
#scatterplot
sns.scatterplot(x='release_year',y='show_id',data=data)
plt.title('Scatterplot')
plt.show()
```

```python
#Hue
sns.scatterplot(x='release_year',y='show_id',data=data,hue='type')
plt.title('Scatterplot')
plt.show()
```
