# Load Dataset

```python
df=sns.load_dataset('titanic')#seaborn ha default datasets
```

```python
df.head()
```

```python
df.columns
```

# Change column name:

```python
df=df.rename(columns={'sex':'gender','class':'class_name'})
```

```python
df.columns
```

#identifying outliers using IQR


```python
import numpy as np
df['fare'].isna().sum()
```

```python
fare=df['fare'].dropna()
```

```python
q1=fare.quantile(0.25)
q3=fare.quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python
outliers=df[(df['fare']<lower_bound) | (df['fare']>upper_bound)]
print("outliers",outliers[['fare']])
```

#creating new columns in exited dataset


```python
df['age_category']=df['age'].apply(lambda x:'senior' if x>=60 else 'adult' if x>=18 else 'child')
print(df[['age','age_category']].head())
df.columns
```

#groupby and aggregation


```python
df_groupby=df.groupby('gender').agg({'fare':'mean','age':'mean'})
print(df_groupby)
```

#pivoting data


```python
df_pivot=df.pivot_table(index='gender',columns='class_name',values='fare',aggfunc='mean')
print(df_pivot)
```

#memory optimization(DownCasting)

```python
df['age']=pd.to_numeric(df['age'],downcast='float')
print(df['age'].dtype)
```

#EDA:-Exploratory Data Analysis

*   It involves analyzing and summarizing the main characteristics of a dataset.
*   EDA can be done using various libraries like Pandas, Matplotlib, Seaborn,
    and NumPy.
*   Encode categorical data, detect outliers, and create new meaningful features for better insights.


```python
!pip install matplotlib
```

```python
!pip install seaborn
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = sns.load_dataset("tips")
a.head()
```

# Basic Plots:
1) Lineplot

*   A line plot is used to visualize trends over time or continuous data.
*   A line plot is used to visualize trends, patterns, and changes in continuous data over time or another sequential variable.





```python
sns.lineplot(x='total_bill',y='tip',data=a)
plt.title("Line plot for Tips dataset")#optional
plt.figure(figsize=(10,5))#optional
plt.show()
```

2) Barplot:


*   A bar plot is used to represent categorical data with rectangular bars.
*   The height of the bars corresponds to the values.



```python
sns.barplot(x='day',y='total_bill',data=a)
plt.title("Bar plot for Tips dataset")
plt.figure(figsize=(10,5))
plt.show()
```

3) Scatter plot:

*   A scatter plot is used to visualize the relationship between two numerical variables by plotting individual data points.
*   It helps identify patterns, correlations, and outliers in datasets.





```python
sns.scatterplot(x='total_bill',y='tip',data=a)
plt.title("Scatter plot for Tips dataset")
plt.figure(figsize=(10,5))
plt.show()
```

Hue :

*   It refers to the color differentiation used to represent different categories in a dataset.
*   It helps in distinguishing groups within the same plot by assigning unique colors.






```python
#hue:
sns.scatterplot(x='total_bill',y='tip',data=a,hue='sex')
plt.title("Scatter plot for Tips dataset")
plt.figure(figsize=(10,5))
plt.show()
```

# UNIVARIATE:
1) Histogram:

*   
It is used in univariate analysis to visualize the distribution of a single numerical variable by dividing data into bins.
*   It helps understand the frequency distribution, skewness, and outliers of a dataset.

Bins:

*   Bins divide the range of data into intervals in a histogram, determining the granularity of the distribution.
*   A small number of bins may oversimplify the data, while too many bins can introduce noise.

KDE (Kernel Density Estimation):

* KDE is a smooth approximation of a histogram that estimates the probability density function of a continuous variable.


```python
sns.histplot(a['total_bill'])
plt.title("Histogram for Tips dataset")
plt.figure(figsize=(10,5))
plt.show()
```

```python
# Histogram using bin size and kde
sns.histplot(a['total_bill'],bins=20,kde=True)
plt.title("Histogram for Tips dataset")
plt.figure(figsize=(10,5))
plt.show()
```

2) Box plot:

*   
It is used in univariate analysis to visualize the distribution, central tendency, and outliers of a single numerical variable.
*   It includes the minimum, first quartile (Q1), median (Q2), third quartile (Q3), maximum, and outliers.


```python
sns.boxplot(a['total_bill'])
plt.title("Box plot for Tips dataset")
plt.figure(figsize=(10,5))
plt.show()
```

3) violin plot:

*   
A violin plot is a combination of a box plot and a kernel density plot, used to visualize the distribution and probability density of a single numerical variable.
*   It provides more insight into the data distribution compared to a box plot.





```python
sns.violinplot(a['total_bill'])
plt.title("Violin plot for Tips dataset")
plt.figure(figsize=(5,5))
plt.show()
```

4) Countplot:

*   
Count Plot shows the number of occurrences of each category in a categorical variable.
*   Visualizes the frequency of categories using bars.

*   Used to identify imbalances or trends in categorical data.
*   Can be customized with hue to differentiate subcategories.



```python
sns.countplot(a['total_bill'])
```

5) Kernal Density Esimator:

*   
Smooths the distribution of a numerical variable to estimate its probability density function.



```python
sns.kdeplot(a['total_bill'])
```
