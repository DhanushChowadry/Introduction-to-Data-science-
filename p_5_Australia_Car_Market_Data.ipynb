# Questions

```python
#What is the average price of cars across different brands, and which brands have the highest and lowest average prices?
data.groupby('Brand')['Price'].mean()
```

```python
#How does the condition of a car (Status) affect its price?
data  = data.rename(columns={'Status': 'Condition'})
sns.barplot(x='Condition', y='Price', data=data)
plt.title('Price vs Condition')
```

```python
#Is there a correlation between mileage (Kilometers) and the price of a car?
data1 = data['Kilometers'].corr(data['Price'])
data1 = pd.DataFrame([[data1]], columns=['Correlation Between Kilometers and Price'])
sns.heatmap(data1, annot=True)
```

```python
#How does engine capacity (CC) relate to car price and mileage?
sns.barplot(x='CC', y='Price', data=data)
plt.title('Price vs CC')
```
