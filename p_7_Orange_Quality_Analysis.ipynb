# Questions

```python
#Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit?
data[["Weight (g)","Size (cm)"]].corr()
sns.scatterplot(x="Weight (g)", y="Size (cm)", data=data)
plt.title("Correlation between Size and Weight")
```

```python
#How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity?
data[["Brix (Sweetness)","pH (Acidity)"]].corr()
```

```python
#How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))?
data[["Softness (1-5)","Ripeness (1-5)"]].corr()
```

```python
#Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels?
data.groupby("Blemishes (Y/N)")[["Softness (1-5)", "Ripeness (1-5)"]].mean()
```

```python
#How do different fruit varieties (Variety) differ in terms of size, weight, sweetness, acidity,andripeness?
data.groupby("Variety")[["Size (cm)", "Weight (g)", "Brix (Sweetness)", "pH (Acidity)", "Ripeness (1-5)"]].mean()
```
